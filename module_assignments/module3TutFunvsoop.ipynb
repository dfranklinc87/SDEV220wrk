Given an array arr[] containing only 0s, 1s, and 2s. Sort the array in ascending order.
Note: You need to solve this problem without utilizing the built-in sort function.

```python
def qsorter(arr):
    if len(arr) <= 1:
       return arr

    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    mid = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return qsorter(left) + mid +qsorter(right)

print(qsorter([2,1,2,2,2,2,1,1,0,0,0,2,1,0]))
```

Given a sorted array arr[] and an integer k, find the position(0-based indexing) at which k is present in the array using binary search. If k doesn't exist in arr[] return -1. 

```python
def bi_search(arr, k):
    first = 0 
    last = len(arr) - 1

    while first <= last:
        pivot = (first + last) // 2

        if arr[pivot] == k:
            return pivot
        elif arr[pivot] < k:
            first = pivot + 1
        else:
            last = pivot - 1

    return -1
    
arr = [1, 3, 3, 5, 7, 9, 11, 13]
k = 11

print(bi_search(arr,k))
        

```
